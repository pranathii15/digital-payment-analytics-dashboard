```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/transactions.csv")

print("Dataset Loaded Successfully!")
print(df.shape)
```

```python
df.info()
```

```python
df.head()
```

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['month'] = df['timestamp'].dt.month_name()

df['quarter'] = df['timestamp'].dt.quarter

df['year'] = df['timestamp'].dt.year

df[['timestamp', 'month', 'quarter', 'year']].head()
```

```python
df['high_value_transaction'] = np.where(df['amount (INR)'] > 5000, 1, 0)
df[['amount (INR)', 'high_value_transaction']].head()
```

```python
df['transaction_status'].value_counts()
```

```python
df['fraud_flag'].value_counts()
```

```python
df.groupby('merchant_category')['amount (INR)'].sum().sort_values(ascending=False)
```

```python
monthly_trend = df.groupby('month')['amount (INR)'].sum()

monthly_trend
```

```python
monthly_trend = df.groupby(
    pd.Categorical(
        df['month'],
        categories=[
            'January','February','March','April','May','June',
            'July','August','September','October','November','December'
        ],
        ordered=True
    )
)['amount (INR)'].sum()

monthly_trend
```

```python
fraud_by_category = df[df['fraud_flag'] == 1].groupby(
    'merchant_category'
)['fraud_flag'].count().sort_values(ascending=False)

fraud_by_category
```

```python
fraud_by_bank = df[df['fraud_flag'] == 1].groupby(
    'sender_bank'
)['fraud_flag'].count().sort_values(ascending=False)

fraud_by_bank
```

```python
fraud_by_state = df[df['fraud_flag'] == 1].groupby(
    'sender_state'
)['fraud_flag'].count().sort_values(ascending=False)

fraud_by_state
```

```python
df['high_value_transaction'].value_counts()
```

```python
df.to_csv('../data/cleaned/transactions_cleaned.csv', index=False)

print("Cleaned dataset exported successfully!")
```
